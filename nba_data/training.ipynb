```python
import pandas as pd
```

```python
df = pd.read_csv('./training.csv', index_col='Unnamed: 0')
```

```python
df
```

```python
df['SEASON'] = df['SEASON_ID'].apply(lambda x: str(x)[1:])
```

```python
import json
```

```python
with open('../api/app/ga/data/team_mapping_id.json') as f:
    mapping = json.load(f)
```

```python
df['TEAM_HOME'] = df['TEAM_NAME_HOME'].apply(lambda x: x.split()[-1])
df['TEAM_AWAY'] = df['TEAM_NAME_AWAY'].apply(lambda x: x.split()[-1])
df['TEAM_HOME'] = df['TEAM_HOME'].map(mapping)
df['TEAM_AWAY'] = df['TEAM_AWAY'].map(mapping)
```

```python
df['TARGET'] = df['WL_HOME'].apply(lambda x: int(x == 'W'))
```

```python
df = df[df['GAME_DATE'] >= '2023-12-01']
```

```python
df[["GAME_DATE", "SEASON", "TEAM_HOME", "TEAM_AWAY", "TARGET"]].to_csv('training_data_2024.csv', index=False)
```
